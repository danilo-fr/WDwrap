```python
def countLines(file):
    linicio = 0
    lfim = 0
    with open(file, "r") as l:
        for number, lines in enumerate(l):
            if re.findall(r'\s*JD\s+Phase', lines):
                linicio=number+1
            if linicio != 0 and re.findall(r'^$', lines):
                lfim = number - linicio
                break
    l.close()
    return linicio, lfim

def countLines2(file):
    linicio = 0
    lfim = 0
    with open(file, "r") as l:
        for number, lines in enumerate(l):
            if re.findall(r'\s*Star\s+M/Msun', lines):
                linicio=number
            if linicio != 0 and re.findall(r'^$', lines):
                lfim = number - linicio
                break
    l.close()
    return linicio, lfim
```

- Define the parameter ranges for MCMC

```python
def log_prior(theta):
    qt, it, ecct, perrt, T2t, phi1t, phi2t, f = theta
    if (q0-0.2) < qt < (q0+0.2) and (i0-1.0) < it < (i0+1.0) and ecc0-0.1 < ecct < ecc0+0.1 and perr0-0.25 < perrt < perr0+0.25 and (T20-0.10) < T2t < (T20+0.10) and (phi10-2.5) < phi1t < (phi10+2.5) and (phi20-2.5) < phi2t < (phi20+2.5) and -9.0 < f < 1.0:
        return 0.0
    return -np.inf
```

```python
def log_probability(theta, x, y, ey):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, x, y, ey)
```

- Definir os valores iniciais (2 x N parametros pelo menos)

```python
pos1 = np.array([])
pos2 = np.array([])
for i in range(len(soln)*2):
    pos = soln[0] + 0.4 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[1] + 2.0 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[2] + 0.2 * 1e-2 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[3] + 0.5 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[4] + 0.2 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[5] + 5.0 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[6] + 5.0 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[7] + 10 * 1e-4 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    
pos2 = np.concatenate((pos2,pos1))
pos3 = np.reshape(pos2, (-1, len(soln)))
nwalkers, ndim = pos3.shape
```

- Run the MCMC

```python
with Pool() as pool:
    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=(x, y, ey), threads=cpu_count()) #, pool=pool)
    sampler.run_mcmc(pos3, 1, progress=True)
    
samples = sampler.chain[:, :, :].reshape((-1, ndim))
#samples = sampler.get_chain(discard=0, thin=1, flat=True)
np.savetxt('samples.dat', samples, fmt='%s')
```
