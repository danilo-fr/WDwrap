```python
def countLines(file):
    linicio = 0
    lfim = 0
    with open(file, "r") as l:
        for number, lines in enumerate(l):
            if re.findall(r'\s*JD\s+Phase', lines):
                linicio=number+1
            if linicio != 0 and re.findall(r'^$', lines):
                lfim = number - linicio
                break
    l.close()
    return linicio, lfim

def countLines2(file):
    linicio = 0
    lfim = 0
    with open(file, "r") as l:
        for number, lines in enumerate(l):
            if re.findall(r'\s*Star\s+M/Msun', lines):
                linicio=number
            if linicio != 0 and re.findall(r'^$', lines):
                lfim = number - linicio
                break
    l.close()
    return linicio, lfim

def countLines3(file):
    linicio = 0
    lfim = 0
    with open(file, "r") as l:
        for number, lines in enumerate(l):
            if re.findall(r'\s*Y\s+Sky', lines):
                linicio=number+1
            if linicio != 0 and re.findall(r'^$', lines):
                lfim = number - linicio
                break
    l.close()
    return linicio, lfim

def countLines4(file):
    linicio = 0
    lfim = 0
    with open(file, "r") as l:
        for number, lines in enumerate(l):
            if re.findall(r'\s*star\s+r pole', lines):
                linicio=number+2
            if linicio != 0 and re.findall(r'^$', lines):
                lfim = linicio+2
                break
    l.close()
    return linicio, lfim
```

- Define the parameter ranges for MCMC

```python
def log_prior(theta):
    it, pot1t, pot2t, f = theta
    if (i0-0.4) < it < (i0+0.4) and (pot10-5.50) < pot1t < (pot10+5.50) and (pot20-2.20) < pot2t < (pot20+2.20) and -9.0 < f < 1.0:
        return 0.0
    return -np.inf
```

```python
def log_probability(theta, phase, y, ey):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, phase, y, ey)
```

- Definir os valores iniciais (2 x N parametros pelo menos)

```python
pos1 = np.array([])
pos2 = np.array([])
for i in range(len(soln)*5):
    pos = soln[0] + 0.80 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[1] + 11.00 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[2] + 4.40 * 1e-3 * np.random.randn(1)
    pos1 = np.append(pos1,pos)
    pos = soln[3] + 10.0 * 1e-4 * np.random.randn(1)
    pos1 = np.append(pos1,pos)

pos2 = np.concatenate((pos2,pos1))
pos3 = np.reshape(pos2, (-1, len(soln)))
nwalkers, ndim = pos3.shape
pos3.shape
```

- Run the MCMC

```python
with Pool() as pool:
    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=(phase, y, ey), threads=cpu_count()) #, pool=pool)
    sampler.run_mcmc(pos3, 1500, progress=True)

samples = sampler.chain[:, :, :].reshape((-1, ndim))
#samples = sampler.get_chain(discard=0, thin=1, flat=True)
np.savetxt('samples.dat', samples, fmt='%s')
```
