```python
def countLines(file):
    linicio = 0
    lfim = 0
    with open(file, "r") as l:
        for number, lines in enumerate(l):
            if re.findall(r'\s*JD\s+Phase', lines):
                linicio=number+1
            if linicio != 0 and re.findall(r'^$', lines):
                lfim = number - linicio
                break
    l.close()
    return linicio, lfim

def countLines2(file):
    linicio = 0
    lfim = 0
    with open(file, "r") as l:
        for number, lines in enumerate(l):
            if re.findall(r'\s*Star\s+M/Msun', lines):
                linicio=number
            if linicio != 0 and re.findall(r'^$', lines):
                lfim = number - linicio
                break
    l.close()
    return linicio, lfim
```

- Define the parameter ranges for MCMC

```python
def log_prior(theta):
    p, ep, q, i, pot1, pot2, T2, flux, f = theta
    if (p0-0.001) < p < (p0+0.001) and (ep0-0.01) < ep < (ep0+0.01) and (q0-0.10) < q < (q0+0.10) and (i0-5.0) < i < (i0+5.0) and (pot10-0.45) < pot1 < (pot10+0.45) and (pot20-0.30) < pot2 < (pot20+0.30) and (T20-0.12) < T2 < (T20+0.12) and (flux0-0.005) < flux < (flux0+0.005) and -9.0 < f < 1.0:
    #if (p0-0.001) < p < (p0+0.001) and (ep0-0.01) < ep < (ep0+0.01) and (q0-0.10) < q < (q0+0.10) and (i0-4.0) < i < (i0+4.0) and (pot10-0.45) < pot1 < (pot10+0.45) and (pot20-0.35) < pot2 < (pot20+0.35) and (T20-0.10) < T2 < (T20+0.10) and (flux0-0.005) < flux < (flux0+0.005) and -9.0 < f < 1.0:
        return 0.0
    return -np.inf
```

```python
def log_probability(theta, x, y, ey):
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    return lp + log_likelihood(theta, x, y, ey)
```

- Run the MCMC

```python
with Pool() as pool:
    sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=(x, y, ey), threads=cpu_count()) #, pool=pool)
    sampler.run_mcmc(pos3, 2223, progress=True)

samples = sampler.chain[:, :, :].reshape((-1, ndim))
#samples = sampler.get_chain(discard=0, thin=1, flat=True)
np.savetxt('samples.dat', samples, fmt='%s')
```

```python
bparam = []
for j in range(ndim):
    mcmc = np.percentile(samples[:, j], [16, 50, 84])
    dif = np.diff(mcmc)
    txt = "\mathrm{{{3}}} = {0:.6f}_{{-{1:.6f}}}^{{{2:.6f}}}"
    txt = txt.format(mcmc[1], dif[0], dif[1], labels[j])
    bparam.append([labels[j], mcmc[1], dif[0], dif[1]])
    display(Math(txt))
```
